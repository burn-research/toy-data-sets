# Data sets composed of hyperplanes - 2D data sets

***

```python
import numpy as np
from PCAfold import center_scale
from PCAfold import plot_2d_manifold
```

***

## Three lines, connected

```python
n_observations = 1000
x = np.linspace(0, 50, n_observations)
phi_1 = np.zeros((n_observations,1))

for observation, x_value in enumerate(x):
    if x_value <= 10:
        phi_1[observation] = 2 * x_value
    elif x_value > 10 and x_value <= 35:
        phi_1[observation] = - x_value + 30
    elif x_value > 35:
        phi_1[observation] = 4 * x_value - 145

data_set_2d = np.hstack((x[:,None], phi_1))

(three_lines_connected, _, _) = center_scale(data_set_2d, scaling='0to1')
plt_handle = plot_2d_manifold(three_lines_connected[:,0], three_lines_connected[:,1], color='k', x_label='$x$', y_label='$y$', figure_size=(10,4), save_filename='hyperplanes-2D-data-sets-three-lines-connected.png')
np.savetxt('hyperplanes-2D-data-sets-three-lines-connected.csv', (three_lines_connected), delimiter=',', fmt='%.16e')
```

## Three lines, disconnected

```python
n_observations = 1000
x = np.linspace(0, 50, n_observations)
phi_1 = np.zeros((n_observations,1))

for observation, x_value in enumerate(x):
    if x_value <= 10:
        phi_1[observation] = 2 * x_value
    elif x_value > 10 and x_value <= 35:
        phi_1[observation] = - x_value 
    elif x_value > 35:
        phi_1[observation] = - 4 * x_value 

data_set_2d = np.hstack((x[:,None], phi_1))

(three_lines_disconnected, _, _) = center_scale(data_set_2d, scaling='0to1')
plt_handle = plot_2d_manifold(three_lines_disconnected[:,0], three_lines_disconnected[:,1], color='k', x_label='$x$', y_label='$y$', figure_size=(10,4), save_filename='hyperplanes-2D-data-sets-three-lines-disconnected.png')
np.savetxt('hyperplanes-2D-data-sets-three-lines-disconnected.csv', (three_lines_disconnected), delimiter=',', fmt='%.16e')
```
